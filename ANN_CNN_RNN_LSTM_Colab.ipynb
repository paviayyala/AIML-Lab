```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd

# ---- Simulated Real-World Sales Data ----
np.random.seed(42)

# advertising spend in thousands ( or $)
ad_spend = np.random.uniform(0, 100, 200).reshape(-1,1)

# sales in thousands with real-world noise
sales = 5 * ad_spend + 20 + np.random.normal(0, 10, ad_spend.shape)

# Convert to float32 for TensorFlow
X = ad_spend.astype(np.float32)
y = sales.astype(np.float32)

# ---- Build ANN Model ----
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=(1,)),
    layers.Dense(8, activation='relu'),
    layers.Dense(1)   # Output = 1 value (sales)
])

#configuring how the neural network learns  like setting rules before training begins
#adam (Adaptive Moment Estimation) -model how to adjust weights during training to reduce error.
#mse (Mean Squared Error)-Loss function = how we measure mistakes
#mae (Mean Absolute Error) -what we monitor during training
#Train using Adam optimizer, measure error with MSE, and show MAE as progress metric.
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# ---- Train Model ----
history = model.fit(X, y, epochs=200, batch_size=16, verbose=0)

# ---- Predict New Sales ---- To VALIDATE THE MODEL
# New advertising spend (in thousands)
new_ad_budget = np.array([[15],[40],[80]], dtype=np.float32)
pred_sales = model.predict(new_ad_budget)

print("Advertising Spend (k):", new_ad_budget.flatten())
print("Predicted Sales (k):\n", pred_sales)
```

## CNN: MNIST Classification

```python
#  Import TensorFlow (Deep Learning library) and Keras (high-level API for building models)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

#  Load the MNIST dataset (handwritten digits: 0-9)
# MNIST already comes with TensorFlow, so no download needed
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

#  Preprocessing the input images
# MNIST images are 28x28 grayscale  shape: (batch, 28, 28)
# Add a channel dimension (..., None)  (batch, 28, 28, 1)
# Normalize pixel values from [0,255]  [0,1] for faster & stable training
x_train = (x_train[..., None] / 255.0).astype("float32")
x_test  = (x_test[..., None]  / 255.0).astype("float32")

#  Build a CNN (Convolutional Neural Network)
model = keras.Sequential([

    #  Input layer  expecting images of size 28x28 with 1 color channel (grayscale)
    layers.Input(shape=(28, 28, 1)),

    #  Convolution layer  detects simple patterns (edges, curves)
    # 32 filters, 3x3 kernel, ReLU activation for non-linearity
    layers.Conv2D(32, 3, activation='relu'),

    #  MaxPooling  reduces image size (downsampling), keeps important features
    layers.MaxPooling2D(),

    #  Another Convolution layer  deeper layer learns more complex patterns (like shapes)
    layers.Conv2D(64, 3, activation='relu'),

    #  More pooling  reduces computation + prevents overfitting
    layers.MaxPooling2D(),

    #  Flatten  convert image feature maps to a 1-D vector for dense layers
    layers.Flatten(),

    #  Fully connected dense layer  learns final classification features
    layers.Dense(64, activation='relu'),

    #  Output layer  10 neurons = 10 digit classes (09)
    # Softmax turns outputs into probabilities that sum to 1
    layers.Dense(10, activation='softmax')
])

#  Compile the model  configure how it learns
model.compile(
    optimizer='adam',                           # Adaptive optimizer (adjusts learning rate smartly)
    loss='sparse_categorical_crossentropy',     # Loss for multi-class classification with integer labels
    metrics=['accuracy']                        # Display accuracy while training
)

#  Train the model
# epochs = how many times we see the entire training dataset
# batch_size = how many images processed at once
# validation_split = 10% of training data used to check performance during training
# verbose=0  don't print training logs (set to 1 to show progress)
model.fit(x_train, y_train, epochs=3, batch_size=128, validation_split=0.1, verbose=0)

#  Test the model on unseen data (x_test)
model.evaluate(x_test, y_test, verbose=2)
```

## RNN: IMDB Sentiment Analysis

This code is building and training a sentiment analysis model on the IMDB movie review dataset using a Simple RNN neural network. It predicts whether a movie review is positive or negative.

```python
# Importing TensorFlow and Keras modules
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define vocabulary size (i.e., number of unique words to consider in the dataset)
vocab_size = 10000  # We will only use the top 10,000 most frequent words

# Define maximum sequence length for each review
max_len = 200  # Each movie review will be padded/truncated to 200 words

# Load the IMDB dataset with only the top 10,000 words kept
# x_train, x_test are sequences of integers (word indices)
# y_train, y_test are labels (0 = negative review, 1 = positive review)

(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data(num_words=vocab_size)

# Pad/truncate sequences so that each review has exactly max_len tokens
# Short reviews get padded with zeros, longer ones are cut
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_len)
x_test  = keras.preprocessing.sequence.pad_sequences(x_test,  maxlen=max_len)

# Build a Sequential model
model = keras.Sequential([

    # Embedding layer converts word indices into dense 64-dimensional vectors
    # input_dim = vocabulary size
    # output_dim = dimension of embedding vector for each word
    # input_length = length of each input sequence (200)
    layers.Embedding(input_dim=vocab_size, output_dim=64, input_length=max_len),

    # SimpleRNN layer processes sequence word-by-word and learns context
    # It outputs a single vector representing meaning of the whole review
    layers.SimpleRNN(64),

    # Dense output layer with sigmoid activation
    # Output = probability between 0 and 1 (binary classification)
    layers.Dense(1, activation='sigmoid')
])

# Compile the model
# optimizer = Adam (adaptive learning optimizer)
# loss = binary cross-entropy (because it's a 0/1 sentiment prediction task)
# metrics = accuracy to measure performance
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
# epochs = number of passes over the dataset
# batch_size = number of samples processed before updating weights
# validation_split = 10% of training data used to validate during training
model.fit(x_train, y_train, epochs=2, batch_size=128, validation_split=0.1)

# Evaluate the model and print clean output
loss, accuracy = model.evaluate(x_test, y_test, verbose=2)

print("\n Model Performance on Test Data")
print(f" Accuracy : {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f" Loss     : {loss:.4f}")
```

## LSTM: IMDB Sentiment Analysis

```python
# Importing TensorFlow and Keras modules required for building neural networks
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define the vocabulary size  i.e., the number of most frequent words we want to keep
# IMDB dataset has many words, but we keep only top 10,000 to limit model size & training time
vocab_size = 10000

# Define maximum length of each movie review (in terms of word tokens)
# Reviews shorter than this will be padded, longer ones will be truncated
max_len = 200

# Load the IMDB dataset
# This dataset returns movie reviews already converted to integer sequences (word indexes)
# x_train, x_test = input review sequences
# y_train, y_test = sentiment labels (0 = negative, 1 = positive)
(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data(num_words=vocab_size)

# Pad sequences so all reviews have the same length (200)
# Padding ensures every review is equal size, which is required for RNN/LSTM models
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_len)
x_test  = keras.preprocessing.sequence.pad_sequences(x_test,  maxlen=max_len)

# Building a Sequential neural network model
model = keras.Sequential([

    # Embedding layer: converts word indices into dense vectors of fixed size (64 dimensions)
    # Acts like creating a lookup table where each word index gets a meaningful vector
    # input_dim = size of vocabulary
    # output_dim = each word will be represented as a 64-dimension vector
    # input_length = each input sample has 200 tokens
    layers.Embedding(input_dim=vocab_size, output_dim=64, input_length=max_len),

    # LSTM layer (Long Short-Term Memory)
    # This layer reads the text sequentially and learns which words to "remember" or "forget"
    # Helps capture long-term context in sentences better than SimpleRNN
    layers.LSTM(64),

    # Fully connected output layer
    # Units = 1  we need one probability score (positive or negative sentiment)
    # Sigmoid converts output to range 01
    layers.Dense(1, activation='sigmoid')
])

# Compile the model  tell it how to learn
# optimizer='adam'  adaptive learning algorithm, good default choice
# loss='binary_crossentropy'  suitable for binary classification (0 or 1)
# metrics=['accuracy']  evaluate performance based on accuracy
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model on training data
# epochs=2  two passes through full dataset
# batch_size=128  process 128 reviews at a time before updating weights
# validation_split=0.1  use 10% of training data as validation set to monitor overfitting
model.fit(x_train, y_train, epochs=2, batch_size=128, validation_split=0.1)

# Evaluate the model on test dataset (unseen data)
# verbose=2  clean evaluation output display
loss, accuracy = model.evaluate(x_test, y_test, verbose=2)

# Print model performance in a friendly format
print("\n Model Performance on Test Data")
print(f" Accuracy : {accuracy:.4f} ({accuracy*100:.2f}%)")  # accuracy in decimal and %
print(f" Loss     : {loss:.4f}")                            # loss value
```
